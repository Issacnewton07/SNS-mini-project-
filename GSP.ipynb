```python
import numpy as np
import networkx as nx

# Load EEG data: matrix (channels x time)
Fs = 256
t = np.linspace(0, 10, 10*Fs)
# 8-channel synthetic EEG with alpha waves + noise
data = np.array([0.5*np.sin(2*np.pi*10*t) + 0.2*np.random.randn(len(t))
                 for _ in range(8)])

# Construct adjacency (here by thresholding Pearson correlation)
corr = np.corrcoef(data)
threshold = 0.5
A = (np.abs(corr) > threshold).astype(int)
G = nx.from_numpy_array(A)

# Compute graph Laplacian L = D - A
L = nx.laplacian_matrix(G).astype(float).todense()

# Graph Fourier basis (eigendecomposition of L)
eigs, U = np.linalg.eigh(L)

# Graph Fourier Transform of first time point
f0 = data[:, 0]          # signal at time 0
gft = U.T @ f0           # spectral coefficients

# Example spectral features: energy in low vs high graph-frequencies
energy_low  = np.sum(gft[:5]**2)
energy_high = np.sum(gft[-5:]**2)
print("Low-freq energy:", energy_low, "High-freq energy:", energy_high)
```
